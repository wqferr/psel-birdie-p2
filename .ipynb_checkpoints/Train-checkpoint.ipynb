```python
import numpy as np
import pandas as pd
```

```python
def stratified_sample(df, col='CATEGORY', n_per_class=2):
    return df.groupby(col, group_keys=False).apply(lambda x: x.sample(min(len(x), n_per_class)))
```

# Carregamento do arquivo de treino

```python
data = pd.read_csv('products.tsv', sep='\t')
data.set_index('ID', inplace=True)
data['SMARTPHONE'] = data.CATEGORY == 'celular-e-smartphone'
```

```python
stratified_sample(data)
```

# Análise de tendências

# Preprocessamento

**Todos** os smartphones dos produtos coletados possuem o título começando com `"Smartphone"` ou `"Celular"`.  
Isso não ocorre no conjunto de dados que se deseja classificar, portanto não é uma boa ideia usar
esse fato como uma *feature*.

```python
print('Frequência das primeiras palavras de cada produto categorizado como celular')
phone_titles = data[data.CATEGORY == 'celular-e-smartphone'].TITLE
first_words = phone_titles.apply(lambda t: t.split()[0])

first_words.value_counts()
```

```python
clean_titles = phone_titles.apply(lambda t: ' '.join(t.split()[1:]))
data.loc[data.CATEGORY == 'celular-e-smartphone', 'TITLE'] = clean_titles
```

```python
stratified_sample(data)
```

# Classificação

## Primeira abordagem

Como primeira tentativa de classificação, decidi por tentar usar uma versão modificada do classificador desenvolvido para a etapa anterior do processo seletivo.  
As features foram modificadas da seguinte maneira:  

1. Remover as features `smart`, `phone` e `celular`.
2. Adicionar a feature `plus`.
3. Adicionar a feature `memoria`.

```python
import re
```

```python
patterns = [
    r'\b[a-z]\d\b', # contém, p.e., "G5", "S9", ...
    r'\b(?:capa|case)\b', # contém a palavra "capa" ou "case"
    r'\bpara\b', # contém a palavra "para" (p.e. "antena para celular")
    r'plus\b', # contém uma palavra que termine com "plus"
    r'\d+ *GB' # contém algo parecido com uma quantidade de memória
]

patterns_re = [re.compile(pat, re.IGNORECASE) for pat in patterns]

# Usado para nomear as colunas do DataFrame
attr_names = ['letra_num', 'capa', 'para', 'plus', 'memoria']
```

```python
def get_attributes(title):
    title_attributes = []
    for pattern in patterns_re:
        if pattern.search(title) is None:
            title_attributes.append(0)
        else:
            title_attributes.append(1)
    return title_attributes
```

```python
def get_row_attributes(dataframe, *, keep=['TITLE']):
    attributes = []
    for i in dataframe.index:
        row_attr = get_attributes(dataframe.loc[i].TITLE)
        attributes.append(row_attr)
    attr_df = pd.DataFrame(attributes)
    attr_df.columns = [f're_{re_name}' for re_name in attr_names]
    for col in keep:
        attr_df.insert(0, col, dataframe[col])
    return attr_df
```

```python
data_attr = get_row_attributes(data, keep=['TITLE', 'CATEGORY', 'SMARTPHONE'])
```

```python
stratified_sample(data_attr)
```
