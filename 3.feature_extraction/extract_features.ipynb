# Carregamento dos dados

```python
import pandas as pd
from smartphone_features import series_to_attr
```

```python
data = pd.read_csv('smartphones.tsv', sep='\t')
data.set_index('ID', inplace=True)
data.head()
```

## Conversão para features
V. `smartphone_features.py`.

```python
attr = series_to_attr(data.TITLE)
attr.head()
```

```python
print(f'Celulares com marcas conhecidas, mas modelos não identificados: {attr[(attr.MODEL.isnull()) & (~attr.BRAND.isnull())].shape[0]}')
attr[(attr.MODEL.isnull()) & (~attr.BRAND.isnull())].head(10)
```

```python
print(f'Celulares com marcas não reconhecidas: {attr[attr.BRAND.isnull()].shape[0]}')
attr[attr.BRAND.isnull()].head(10)
```

```python
print(f'Total de modelos não identificados: {attr.MODEL.isnull().sum()} de {data.shape[0]}')
```

```python
attr.to_csv('features.tsv', sep='\t', na_rep='N/A')
```
